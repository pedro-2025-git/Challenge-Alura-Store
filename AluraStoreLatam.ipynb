### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt
sumaTienda = tienda['Precio'].sum()
sumaTienda2 = tienda2['Precio'].sum()
sumaTienda3 = tienda3['Precio'].sum()
sumaTienda4 = tienda4['Precio'].sum()

lista = []
lista = ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4']
ingreso = [sumaTienda, sumaTienda2, sumaTienda3, sumaTienda4]

print(f'Ingreso total de la tienda 1 es: $ {sumaTienda}')
print(f'Ingreso total de la tienda 2 es: $ {sumaTienda2}')
print(f'Ingreso total de la tienda 3 es: $ {sumaTienda3}')
print(f'Ingreso total de la tienda 4 es: $ {sumaTienda4}')


```

```python
import matplotlib.pyplot as plt
# Gráfico de Barras para todas las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresosTienda = [sumaTienda, sumaTienda2, sumaTienda3, sumaTienda4]
colores = ['#FF0000', '#0000FF', '#00FF00', '#FFFF00']

plt.figure(figsize=(8, 5))
barras = plt.bar(tiendas, ingresosTienda, width=0.6, color=colores)
plt.title('Facturaciones Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos(Facturación)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# # Agrega etiquetas al gráfico de tiendas
for barra in barras:
    valory = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, valory, f"${valory:,.0f}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Obtener datos agrupados de la tienda 1
df_tienda = tienda.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(f'Ventas por Categoría Tienda 1')
print(df_tienda)
print('\n')

# Obtener datos agrupados de la tienda 2
df_tienda2 = tienda2.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(f'Ventas por Categoría Tienda 2')
print(df_tienda2)
print('\n')

# Obtener datos agrupados de la tienda 3
df_tienda3 = tienda3.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(f'Ventas por Categoría Tienda 3')
print(df_tienda3)
print('\n')

# Obtener datos agrupados de la tienda 4
df_tienda4 = tienda4.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
print(f'Ventas por Categoría Tienda 4')
print(df_tienda4)
print('\n')

```

```python
# Gráfico de barras Tienda 1
import matplotlib
plt.figure(figsize=(6, 3))
barras = plt.barh(df_tienda.index, df_tienda.values, color='Blue')
plt.title('Ventas por Categoría Tienda 1')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.grid(axis='x', linestyle='--', alpha=0.8)
plt.gca().invert_yaxis()  # Ordena las categorias
# Agrega etiquetas Tienda 1
for barra in barras:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

# Gráfico de barras Tienda 2
plt.figure(figsize=(6, 3))
barrasTienda2 = plt.barh(df_tienda2.index, df_tienda2.values, color='Red')
plt.title('Ventas por Categoría Tienda 2')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.grid(axis='x', linestyle='--', alpha=0.8)
plt.gca().invert_yaxis()
# Agrega etiquetas Tienda 2
for barra in barrasTienda2:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

# Gráfico de barras Tienda 3
plt.figure(figsize=(6, 3))
barrasTienda3 = plt.barh(df_tienda3.index, df_tienda3.values, color='Green')
plt.title('Ventas por Categoría Tienda 3')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.grid(axis='x', linestyle='--', alpha=0.8)
plt.gca().invert_yaxis()
# Agrega etiquetas Tienda 3
for barra in barrasTienda3:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

# Gráfico de barras Tienda 4
plt.figure(figsize=(7.4, 3))
barrasTienda4 = plt.barh(df_tienda4.index, df_tienda4.values, color='Brown')
plt.title('Ventas por Categoría Tienda 4')
plt.xlabel('Ventas Totales')
plt.ylabel('Categoría')
plt.grid(axis='x', linestyle='--', alpha=0.8)
plt.gca().invert_yaxis()
# Agrega etiquetas Tienda 4
for barra in barrasTienda4:
    valorx = barra.get_width()
    plt.text(valorx, barra.get_y() + barra.get_height()/2, f"{valorx:,.0f}", va='center', fontsize=8)

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
import numpy as np
promedioTienda = tienda['Calificación'].mean()
promedioTienda2 = tienda2['Calificación'].mean()
promedioTienda3 = tienda3['Calificación'].mean()
promedioTienda4 = tienda4['Calificación'].mean()

print(f'Calificación promedio tienda 1: {promedioTienda:.2f}')
print(f'Calificación promedio tienda 2: {promedioTienda2:.2f}')
print(f'Calificación promedio tienda 3: {promedioTienda3:.2f}')
print(f'Calificación promedio tienda 4: {promedioTienda4:.2f}')

```

```python
# Grafico de puntos por tienda
ejex = np.array(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
ejey = np.array([promedioTienda, promedioTienda2, promedioTienda3, promedioTienda4])
plt.scatter(ejex, ejey)
plt.xlabel('Tiendas')
plt.ylabel('Calificación')
plt.show()
```

# 5. Envío promedio por tienda

```python
import matplotlib.pyplot as plt

costoEnvioTienda = tienda['Costo de envío'].mean()
costoEnvioTienda2 = tienda2['Costo de envío'].mean()
costoEnvioTienda3 = tienda3['Costo de envío'].mean()
costoEnvioTienda4 = tienda4['Costo de envío'].mean()

print('Costo de envío de la tienda 1: ', costoEnvioTienda)
print('Costo de envío de la tienda 2: ', costoEnvioTienda2)
print('Costo de envío de la tienda 3: ', costoEnvioTienda3)
print('Costo de envío de la tienda 4: ', costoEnvioTienda4)

```

```python
# Gráfico de envíos promedio por tienda
listaTiendas = ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4']
costosEnvios = [costoEnvioTienda, costoEnvioTienda2, costoEnvioTienda3, costoEnvioTienda4]
colores = ['#FF0000', '#0000FF', '#00FF00', '#BF00BF']
plt.figure(figsize=(8, 5))
barras = plt.bar(listaTiendas, costosEnvios, width=0.6, color=colores)
plt.title('Costos de envíos promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Promedio de envíos')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Agrega etiquetas as barras
for barra in barras:
    valory = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, valory, f"${valory:,.0f}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
